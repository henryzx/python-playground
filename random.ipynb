```python
import numpy as np
import plotly.express as px

coins_num = 50
times = 100

head_counts = []
for i in range(times):
    results = ['head' if r < 0.5 else 'tail' for r in np.random.rand(coins_num)]
    head_counts.append(results.count('head'))

fig = px.histogram(head_counts)
fig.show()
```
