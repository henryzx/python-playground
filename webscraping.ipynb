```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from download_file import download_file

page_url = 'https://e-hentai.org/s/171fff4735/3613055-1'
page_number = 1

driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.minimize_window()
wait = WebDriverWait(driver, 10)

try:
    while True:
        driver.get(page_url)

        anchor_el = driver.find_element(by=By.CSS_SELECTOR, value='#i3 > a')
        image_el = driver.find_element(by=By.CSS_SELECTOR, value='#img')

        image_url = image_el.get_dom_attribute('src')
        download_file(image_url, f'{page_number}')
        # next page
        next_page_url = anchor_el.get_dom_attribute('href')
        if next_page_url == page_url: 
            break
        else:
            page_url = next_page_url
            page_number += 1
except Exception as e:
    raise e
finally: driver.quit()
```
